### cifar 10

```python
(_, __), (X_test, y_test) = tf.keras.datasets.cifar10.load_data()
y_test=to_categorical(y_test, num_classes=10)
```

```python
SE_resnet = tf.keras.models.load_model('./full_models/cifar10/SE_resnet_best.h5')
```

### 5 first labels

```python
rows=5
cols = 5
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))
for i in range(rows):
    for j in range(cols):        
        axes[i, j].imshow(X_class[i][j])
label_to_name={0:'plane',1:'automobile',2:'bird',3:'cat',4:"deer" }        
```

# Where to retrieve the values

```python
# layers that give excitation values
l=[l.name for l in SE_resnet.layers]
list_se=[]
index_to_layer={}
for i,name in enumerate(l):
    if('SE_fc2' in name):
        list_se.append(i)
        index_to_layer[i]=name
index_to_layer
```

# Take 50 samples for each of the 5 first labels

```python
# data per label
num_class=5
X_class=[]
for k in range(min(10,num_class)):  
    X_class.append([x for i,x in enumerate(X_test) if y_test[i][k]==1][:50])
```

# Plot the activation values per channel for each labal

```python
from tensorflow.keras import backend as K


    
for layer_num in list_se:
    plt.figure(figsize=(8, 8))
    plt.subplot(2, 1, 1)

    for i in range(num_class):
        mean=[]
        for x in X_class[i]:
            get_relu_output = K.function([SE_resnet.input], [SE_resnet.layers[layer_num].output])
            excitations = np.array(get_relu_output(np.array([x]))[0][0]) 
           
            mean.append(excitations)
            
        mean = np.mean(mean, axis=0)
        
      
        plt.plot(mean, label=label_to_name[i],alpha=0.7)

   
    plt.legend(loc='lower right')
    plt.ylabel('Activation')
    plt.ylim([min(plt.ylim()),1])
    plt.title(index_to_layer[layer_num][:-7])
    plt.xlabel('Channel index')
    plt.show()
```

## With standard deviations

```python
for layer_num in list_se:
    plt.figure(figsize=(8, 8))
    plt.subplot(2, 1, 1)
    for i in [1,3]:
        samples=[]
        for x in X_class[i]:
            get_relu_output = K.function([SE_resnet.input], [SE_resnet.layers[layer_num].output])
            excitations = np.array(get_relu_output(np.array([x]))[0][0]) 
           
            samples.append(excitations)
        std= np.std(samples, axis=0)
        mean = np.mean(samples, axis=0)
             
        plt.plot(mean, label=label_to_name[i])
        x = np.linspace(0, len(mean),  len(mean))
        plt.fill_between(x,mean-std,mean+std, alpha=0.3)

       
    plt.legend(loc='lower right')
    plt.ylabel('Activation')
    plt.ylim([min(plt.ylim()),1])
    plt.title(index_to_layer[layer_num][:-7])
    plt.xlabel('Channel index')
    plt.show()
```
